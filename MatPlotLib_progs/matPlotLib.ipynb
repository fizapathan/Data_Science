```python
import matplotlib.pyplot as plt
```

# pie chart

```python
lable = 'egitarian', 'very veg' , 'non veg'
```

```python
size = [1, 1, 4]
```

```python
explode = [0, 0, 0.1]
fig,ax = plt.subplots()
ax.pie(size, labels=lable, startangle=90, autopct='%1.1f%%', shadow=True, explode = explode)
ax.axis('equal')
plt.title("Pi chart")
plt.show()
fig.savefig('pie_chart.png')
```

# scatter plot

```python
import numpy as np

x = np.random.randint(1, 10, 10)
y = np.random.randint(1, 10, 10)
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("x value")
plt.ylabel("y lable")
plt.show()
fig.savefig('scatter.png')
```

# histogram

```python
x = []
for i in range(1, 1000):
    temp = np.random.randint(1, 6, 10)
    temp = sum(temp)
    x.append(temp)
```

```python
fig,ax = plt.subplots()
ax.hist(x)
plt.title("Histogram")
plt.xlabel("x value")
plt.ylabel("y lable")
plt.show()
fig.savefig('hist.png')
print(x)
```

```python
x = np.random.randint(0,100,20)
```

```python
plt.hist(x,5,facecolor ='blue',alpha = 0.5)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar([2,4,6,8], [10, 20, 30, 40], color='b', width=0.5, alpha = 0.6, yerr=[2,1,2,1])
ax.bar([2,4,6,8], [2, 6, 4, 8], color='r', alpha = 0.8, yerr=[2,1,2,1])
ax.bar([1,3,5,7], [5, 10, 15, 20], color = 'y', alpha = 1, yerr=[1,2,1,1])
plt.title("Bar")
plt.xlabel("x value")
plt.ylabel("y lable")
plt.show()
fig.savefig('bar.png')
```

```python
fig,ax = plt.subplots()
ax.barh([1,2,3,4], [10, 20, 30, 40], color='b', alpha = 0.6, xerr=[2,1,2,1])
ax.barh([1,2,3,4], [2, 6, 4, 8], color='r', alpha = 0.8, xerr=[2,1,2,1])
plt.title("Bar")
plt.xlabel("x value")
plt.ylabel("y lable")
plt.show()
fig.savefig('bar.png')
```

```python
fig, ax = plt.subplots()
ax.broken_barh([(110, 30), (150, 10)], (10, 9), facecolors='tab:blue', hatch="+")
ax.broken_barh([(10, 50), (100, 20), (130, 10)], (20, 9),
               facecolors=('tab:orange', 'tab:green', 'tab:red'), hatch="O")
plt.xlabel("x value")
plt.ylabel("y value")
plt.xlim(10, 200)
plt.ylim(10, 30)
ax.set_yticks([15,25])
ax.set_yticklabels(['Y_1', 'Y_2'])
ax.grid(True)
plt.show()
```
