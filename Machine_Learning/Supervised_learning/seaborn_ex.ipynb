```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
sns.barplot(x=[1, 2, 3, 4, 5], y=[20, 30, 50, 80, 100], palette = 'rocket')#default, rocket, vlag
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
sns.barplot(x=[1, 2, 3, 4, 5], y=[-20, -30, 50, 80, 100], palette = 'vlag')#default, rocket, vlag
plt.show()
```

# catplot

```python
fig, ax = plt.subplots(figsize = (10, 5))
sns.catplot([1, 2, 3, 4], [0, 1, 1, 4])#default, rocket, vlag
plt.show()
```
