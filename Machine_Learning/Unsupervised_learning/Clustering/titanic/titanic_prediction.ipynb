```python
import numpy as np
```

```python
x = np.array(data.drop(columns = ['Survived'], inplace = False).astype(float))
```

```python
x
```

```python
y = np.array(data['Survived'])
```

```python
y
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(x)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, kmeans.predict(x))
```

```python
correct = 1
for i in range(len(x)):
    predict = np.array(x[i].astype(float))
    print("Getting from array: ", predict)
    predict = predict.reshape(-1, len(predict))
    prediction = kmeans.predict(predict)
    if prediction[0]==y[i]:
        correct += 1
print(correct/len(x)*100)
```
